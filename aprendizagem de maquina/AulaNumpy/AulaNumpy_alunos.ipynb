## Multiplicação de Matrizes

Para matrizes $\boldsymbol{X} \in \mathbb{R}^{a\times b}$ e $\boldsymbol{Y} \in \mathbb{R}^{b\times c}$, temos que
$\boldsymbol{XY} \in \mathbb{R}^{a\times c}$. No Numpy, essa operação é representada por $\texttt{X @ Y}$.

A multiplicação é definida por:
$$\begin{aligned}
(\boldsymbol{XY})_{ij} = \sum_k \boldsymbol{X}_{ik}\boldsymbol{Y}_{kj}
\end{aligned}$$

````python
print(A @ B)
print("Dimensão de A @ B:", (A@B).shape)
````

### Computar a fórmula da normalização escore-Z

Dado um conjunto de dados $\boldsymbol{X} = [\boldsymbol{x}_1, \boldsymbol{x}_1, \ldots, \boldsymbol{x}_N]$, a normalização por escore-Z é dada por:
$$\tilde{\boldsymbol{x_i}} = \frac{\boldsymbol{x_i} - \boldsymbol{\mu}}{\boldsymbol{\sigma}}$$

Onde:
$$\begin{aligned}
\boldsymbol{\mu} &= \frac{1}{N}\sum_i^N \boldsymbol{x_i}\\
\boldsymbol{\sigma} &= \sqrt{\frac{1}{N-1}\sum_i^N (\boldsymbol{x_i}-\boldsymbol{\mu})^2}\\
\end{aligned}$$

A "desnormalização" pode ser feita por:
$$\boldsymbol{x_i} = \boldsymbol{\sigma}\tilde{\boldsymbol{x_i}} + \boldsymbol{\mu}$$

Use a fórmula para normalizar o conjunto de dados `peixe` sem usar nenhum `for` ou `while`.

````python
# Escreva o código aqui
def normalizar(dados):
    media = np.mean(dados, axis=0)

    matriz_de_desvios = (dados - media) ** 2

    variancia = np.sum(matriz_de_desvios, axis=0) / (dados.shape[0]-1)

    desvio_padrao = np.sqrt(variancia)

    x_normalizado = (dados - media) / desvio_padrao
    
    return x_normalizado, media, desvio_padrao

def desnormalizar(dados_normalizados, media, desvio_padrao):
    return (desvio_padrao*dados_normalizados) + media
    
````

### Encontrar raízes de funções

Dada uma função $f$, o seguinte procedimento (método de Newton-Raphson) consegue encontrar aproximar zeros desta função:

1. Inicialize $\tilde{x}_0$ com um chute inicial e escolha uma tolerância $\epsilon$;
2. Compute: $$\tilde{x}_{t} = \tilde{x}_{t-1} - \frac{f(\tilde{x}_{t-1})}{f'(\tilde{x}_{t-1})}$$
3. Repita o passo 2 até que $f(\tilde{x}_t) \leq \epsilon$.

Implemente esse procedimento para $f(x) = x^2 - 2$, teste seu resultado computando $\tilde{x} \cdot \tilde{x}$.

````python
# Escreva o código aqui
def funcao(x):
    return x**2 - 2

def funcao_derivada(x):
    return 2*x 

n_iteracoes = 20
x = np.zeros(n_iteracoes)
x[0] = 10

for i in range(1, n_iteracoes):
    x[i] = x[i-1] - (funcao(x[i-1]) / funcao_derivada(x[i-1]))
    
print(x[n_iteracoes-1])
````

# Regressão linear via mínimos quadrados ordinários (OLS)

Como visto na aula de regressão linear, dado o seguinte problema:
$$\begin{aligned}
\hat{\boldsymbol{y}} = \boldsymbol{X}\boldsymbol{w}\\
\text{Queremos encontrar:}\\
\hat{\boldsymbol{w}} = \arg\min_{\boldsymbol{w}} ||\boldsymbol{y} - \boldsymbol{X}\boldsymbol{w}||^2
\end{aligned}$$

Como visto, sabemos que:
$$\hat{\boldsymbol{w}} = (\boldsymbol{X}^{\top}\boldsymbol{X})^{-1} \boldsymbol{X}^{\top}\boldsymbol{y}$$

Obtenha um modelo linear para predizer o comprimento de um peixe a partir de sua idade e da temperatura da água. Para isso, encontre o $\hat{\boldsymbol{w}}$ pro dataset `peixe` onde:
$$\boldsymbol{X} = [\mathbf{1}^{\top} \;,\; \mathtt{peixe\underline{}dataset[:,[0,1]}]\\
\boldsymbol{y} = \mathtt{peixe\underline{}dataset[:,[2]]}$$

Sem utilizar estrutura de repetição alguma, calcule a raíz do erro quadrático médio:
$$\mathrm{RMSE} = \sqrt{\frac{1}{n}\sum_i^n (\boldsymbol{y} - \hat{\boldsymbol{y}})^2}$$

````python
# Escreva o código aqui
n = peixe_dataset.shape[0]

ones = np.ones((n, 1))
dados_x = peixe_dataset[:,[0,1]]

X = np.hstack((ones, dados_x))
Y = peixe_dataset[:,[2]]

# Calcular os pesos 
w = (np.linalg.inv((X.T@X))) @ X.T @ Y

# Calculando RMSE
rmse = np.sqrt((((Y - (X@w)).T @ (Y - (X@w))) / n))

print(rmse)
````

# Regressão linear via gradiente descendente (GD)

Sendo $\alpha$ um passo de aprendizagem, $N$ o número de observações disponíveis e $t$ a iteração atual do algoritmo, os parâmetros $\boldsymbol{w}$ podem ser atualizados via algoritmo GD:

$$\boldsymbol{w}(t) = \boldsymbol{w}(t-1) + \alpha \frac{1}{N}\sum_{i=1}^{N} e_i(t-1) \boldsymbol{x}_i,$$
$$\text{em que } e_i(t-1) = y_i - \boldsymbol{w}^{\top}(t-1)\boldsymbol{x}_i$$

Normalize os dados retirando a média e dividindo pelo desvio padrão (entrada e saída). Na predição, desfaça a normalização para computar o RMSE.

````python
# Escreva o código aqui
dados_normalizados, media, desvio_padrao = normalizar(peixe_dataset)
X = np.hstack((np.ones((n, 1)), dados_normalizados[:, [0,1]]))
Y = dados_normalizados[:, [2]]

n = peixe_dataset.shape[0]
n_iteracoes = 200
alfa = 0.1
k = X.shape[1]

w = np.zeros((n_iteracoes, k))

for i in range(1, n_iteracoes):
    # W anterior em coluna
    w_anterior = w[i-1].reshape(k, 1)
    
    Y_predito = X @ w_anterior
    
    erro = Y - Y_predito
    
    gradiente = (X.T @ erro) / n
    
    w_novo = w_anterior + (alfa * gradiente)
    
    w[i] = w_novo.T

print(w[-1])
````

````python
import matplotlib.pyplot as plt

# Vamos calcular o erro em cada iteração para ver o gráfico de convergência
erros_hist = []
for i in range(n_iteracoes):
    pred = X @ w[i].reshape(-1, 1)
    erro_medio = np.mean((Y - pred)**2)
    erros_hist.append(erro_medio)

plt.plot(erros_hist)
plt.title("Convergência do Gradiente Descendente")
plt.xlabel("Iteração")
plt.ylabel("Erro Quadrático Médio")
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Coluna 0 de X é bias (1); só as colunas 1 e 2 são idade e temperatura normalizadas.
# Y é comprimento normalizado (uma coluna): usar media[2] e desvio_padrao[2] apenas.
X_real = np.hstack(
    (np.ones((X.shape[0], 1)), desnormalizar(X[:, 1:], media[:2], desvio_padrao[:2]))
)
Y_real = desnormalizar(Y, media[2:3], desvio_padrao[2:3])

# Definindo a figura com 1 linha e 2 colunas
fig = plt.figure(figsize=(15, 6))
w_final = w[-1]

# --- LADO ESQUERDO: GRÁFICO 2D (Idade vs. Comprimento) ---
ax1 = fig.add_subplot(1, 2, 1)

# Pontos reais (Idade na col 1, Comprimento em Y)
ax1.scatter(X_real[:, 1], Y_real.ravel(), color='blue', alpha=0.5, label='Dados Reais')

# w foi estimado com dados normalizados: predizer em escala normalizada e desnormalizar o comprimento
x_reta = np.linspace(X_real[:, 1].min(), X_real[:, 1].max(), 100)
temp_media = np.mean(X_real[:, 2])
x_reta_norm = (x_reta - media[0]) / desvio_padrao[0]
temp_norm_fixo = (temp_media - media[1]) / desvio_padrao[1]
y_reta_norm = w_final[0] + w_final[1] * x_reta_norm + w_final[2] * temp_norm_fixo
y_reta = y_reta_norm * desvio_padrao[2] + media[2]

ax1.plot(x_reta, y_reta, color='red', linewidth=3, label='Reta de Regressão')
ax1.set_xlabel('Idade')
ax1.set_ylabel('Comprimento')
ax1.set_title('Visão 2D: Idade vs. Comprimento (temp. média fixa)')
ax1.legend()
ax1.grid(True, linestyle='--', alpha=0.7)

# --- LADO DIREITO: GRÁFICO 3D (Plano de Regressão) ---
ax2 = fig.add_subplot(1, 2, 2, projection='3d')

# Pontos reais no espaço 3D
ax2.scatter(X_real[:, 1], X_real[:, 2], Y_real.ravel(), color='blue', alpha=0.5)

# Criando a malha (grade) para o plano
x_range = np.linspace(X_real[:, 1].min(), X_real[:, 1].max(), 20)
y_range = np.linspace(X_real[:, 2].min(), X_real[:, 2].max(), 20)
x_surf, y_surf = np.meshgrid(x_range, y_range)

# Plano no espaço normalizado, depois comprimento em escala original
x_surf_norm = (x_surf - media[0]) / desvio_padrao[0]
y_surf_norm = (y_surf - media[1]) / desvio_padrao[1]
z_surf_norm = w_final[0] + w_final[1] * x_surf_norm + w_final[2] * y_surf_norm
z_surf = z_surf_norm * desvio_padrao[2] + media[2]

# Plotando o plano (superfície)
ax2.plot_surface(x_surf, y_surf, z_surf, color='red', alpha=0.3, cmap='Reds')

ax2.set_xlabel('Idade')
ax2.set_ylabel('Temperatura')
ax2.set_zlabel('Comprimento')
ax2.set_title('Visão 3D: Plano de Regressão')

# Ajusta o layout para não sobrepor os títulos
plt.tight_layout()
plt.show()
````

# Regressão linear via gradiente descendente estocástico (SGD)

Sendo $\alpha$ um passo de aprendizagem, $N$ o número de observações disponíveis e $t$ a iteração atual do algoritmo, os parâmetros $\boldsymbol{w}$ podem ser atualizados via algoritmo SGD:

$$\boldsymbol{w}(t) = \boldsymbol{w}(t-1) + \alpha e_i(t-1) \boldsymbol{x}_i,$$
$$\text{em que } e_i(t-1) = y_i - \boldsymbol{w}^{\top}(t-1)\boldsymbol{x}_i$$

Normalize os dados retirando a média e dividindo pelo desvio padrão (entrada e saída). Na predição, desfaça a normalização para computar o RMSE.

````python
dados_normalizados, media, desvio_padrao = normalizar(peixe_dataset)
n = peixe_dataset.shape[0]
X = np.hstack((np.ones((n, 1)), dados_normalizados[:, [0, 1]]))
Y = dados_normalizados[:, [2]]

n_iteracoes = 200
alfa = 0.1
k = X.shape[1]

w = np.zeros((n_iteracoes, k))

for epoca in range(1, n_iteracoes):
    w_anterior = w[epoca - 1].reshape(k, 1)
    for i in np.random.permutation(n):
        xi = X[i : i + 1, :]
        erro_i = Y[i, 0] - (xi @ w_anterior)[0, 0]
        w_anterior = w_anterior + alfa * erro_i * xi.T
    w[epoca] = w_anterior.ravel()

print(w[-1])

w_final = w[-1].reshape(k, 1)
pred_norm = X @ w_final
Y_pred_real = pred_norm * desvio_padrao[2] + media[2]
Y_real = Y * desvio_padrao[2] + media[2]
rmse = np.sqrt(np.mean((Y_real - Y_pred_real) ** 2))
print("RMSE (comprimento, escala original):", rmse)
````
